```python
# depth issue .. Solved by prespective
import cv2
image = io.imread('test_cases/5a_shadow.jpeg')
pts1 = np.float32([[0, 0], [image.shape[1], 0], [0, image.shape[0]], [image.shape[1], image.shape[0]]])
pts2 = np.float32([[0, 0], [image.shape[1]*2, 0], [0, image.shape[0]*2], [image.shape[1]*2, image.shape[0]*2]])
matrix = cv2.getPerspectiveTransform(pts1, pts2)
result = cv2.warpPerspective(image, matrix, (image.shape[1], image.shape[0]))
show_images([image,result])
io.imsave("test_cases/5a_shear.jpeg",np.array((result*255).astype(np.uint8)))
```
